1. What is NumPy, and why is it widely used in Python ?
   - What is NumPy?
      - NumPy (Numerical Python) is a powerful Python library used for numerical computations. It provides support for multi-dimensional arrays and mathematical functions to operate on them efficiently.

  - Why is NumPy Widely Used?
    - Fast Performance – Uses optimized C and Fortran code for computations.
    - Multi-Dimensional Arrays – Supports ndarray, which is more efficient than Python lists.
    - Mathematical Operations – Provides functions for linear algebra, statistics, and more.
    - Broadcasting – Enables element-wise operations without explicit loops.
    - Memory Efficient – Uses less memory than Python lists.
    - Integration with Other Libraries – Works well with Pandas, SciPy, Matplotlib, etc.
    - Indexing & Slicing – Allows easy manipulation of array elements.

2. How does broadcasting work in NumPy ?
    - Broadcasting in NumPy allows operations on arrays of different shapes without explicitly reshaping them. It automatically expands smaller arrays to match the shape of larger ones.
    - Broadcasting Rules
      - Match dimensions from the right – If dimensions differ, NumPy tries to stretch the smaller one.
      - Size 1 can be stretched – If a dimension is 1, it expands to match the other array.
      - Mismatch in non-1 dimensions leads to an error.

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(5, 5)
correlation_matrix = np.corrcoef(data)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
#8. Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='Sin(x)')
plt.plot(x, y2, label='Cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin(x) and Cos(x)')
plt.legend()
plt.show()
```

```python
#11. Perform matrix multiplication using NumPy
import numpy as np
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
result1 = np.dot(A, B)
result2 = A @ B
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication (np.dot()):\n", result1)
print("Matrix Multiplication (@ operator):\n", result2)
```

```python
#13. Create a 3D scatter plot using Plotly.
import plotly.graph_objects as go
import numpy as np
np.random.seed(42)
x = np.random.rand(20) * 10
y = np.random.rand(20) * 10
z = np.random.rand(20) * 10

fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z, mode='markers',
    marker=dict(size=8, color=z, colorscale='Viridis', opacity=0.8)
)])

fig.update_layout(title="3D Scatter Plot",
                  scene=dict(xaxis_title="X Axis",
                             yaxis_title="Y Axis",
                             zaxis_title="Z Axis"))
fig.show()
```
