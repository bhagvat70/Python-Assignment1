1. What is a random variable in probability theory?
  - A random variable in probability theory is a function that assigns numerical values to the outcomes of a random experiment.
2. What are the types of random variables?
    - Discrete Random Variable ‚Äì Takes a countable number of distinct values (e.g., number of heads in 3 coin tosses).
    - Continuous Random Variable ‚Äì Takes an infinite number of values within a range (e.g., height of people, time taken to finish a race).
3. What is the difference between discrete and continuous distributions?
  - Discrete Distribution
    - Deals with countable values
    - Finite or infinite but distinct values (e.g., 0, 1, 2‚Ä¶)
    - Probability Mass Function (PMF)
    - P(X = x) gives the probability of an exact value

  - Continuous Distribution
    - Deals with uncountable values (range)
    - Any value within an interval (e.g., 1.2, 2.5, 3.8‚Ä¶)
    - Probability Density Function (PDF)
    - P(a ‚â§ X ‚â§ b) is found using integration (area under curve)
4. What are probability distribution functions (PDF)?

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

    - Gives probability at an exact value.
    - If X is a height following a normal distribution, ùëì(ùë•)f(x) gives density at X
  
  - Cumulative Distribution Function (CDF)
    - Describes the probability that a random variable X is less than or equal to a value (P(X ‚â§ x)).
    - Always increases from 0 to 1.
    - Gives cumulative probability up to a point.
    - F(170)=P(X‚â§170), found by integrating PDF from ‚àí‚àû  to 170.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original population distribution.
  - Why Is CLT Critical in Statistics?
     - Enables Inferential Statistics
          - Allows us to make predictions about a population using a sample.
     - Foundation for Hypothesis Testing & Confidence Intervals
          - Used in Z-tests, t-tests, and confidence interval estimation.
     - Real-World Applications
          - Polling: Estimating population preferences from survey samples.
          - Quality Control: Checking manufacturing defects from random samples.
          - Finance: Modeling stock returns using sample data.
14. How does the Central Limit Theorem relate to the normal distribution?
  - The Central Limit Theorem (CLT) explains why the normal distribution appears in many real-world scenarios, even when the original data is not normally distributed. It establishes a direct link between sampling distributions and the normal distribution.

```python
#.3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Compute the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The random variable (0 or 1)
    p (float): The probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: The probability mass function (PMF) value for the given x
    """
    if x not in [0, 1]:
        return 0
    return p**x * (1 - p)**(1 - x)
p = 0.6
print("P(X=0):", bernoulli_pdf(0, p))
print("P(X=1):", bernoulli_pdf(1, p))
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    return (data - mean) / std_dev

data = np.random.normal(50, 15, 100)
z_scores = calculate_z_scores(data)

plt.figure(figsize=(8, 5))
sns.histplot(z_scores, bins=30, kde=True, color='blue')
plt.axvline(0, color='red', linestyle='dashed', label="Mean (Z=0)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Z-score Distribution")
plt.legend()
plt.grid()
plt.show()
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = population_std / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    reject_null = p_value < alpha
    return z_score, p_value, reject_null

sample_data = np.random.normal(52, 10, 30)
population_mean = 50
population_std = 10
alpha = 0.05

z_score, p_value, reject_null = z_test(sample_data, population_mean, population_std, alpha)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print("Reject Null Hypothesis" if reject_null else "Fail to Reject Null Hypothesis")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    margin_of_error = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1) * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

data = np.random.normal(50, 10, 30)
conf_int = confidence_interval(data, 0.95)

print(f"95% Confidence Interval: ({conf_int[0]:.2f}, {conf_int[1]:.2f})")

```

```python
  #17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    margin_of_error = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1) * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean

data = np.random.normal(50, 10, 30)
lower, upper, mean = confidence_interval(data, 0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

if lower <= 50 <= upper:
    print("The population mean of 50 is within the confidence interval.")
else:
    print("The population mean of 50 is NOT within the confidence interval.")

```
